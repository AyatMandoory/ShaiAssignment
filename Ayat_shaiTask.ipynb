```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/Salaries.csv')
df.head()

num_of_rows, num_of_cols = df.shape
print(f"Number of rows: {num_of_rows}")
print(f"Number of columns: {num_of_cols}")

data_types = df.dtypes
print("\nData types of each column:")
print(data_types)

missing_values = df.isnull().sum()
print("\nMissing values in each column:")
print(missing_values)
```

```python
salaries = df['TotalPay']
mean =salaries.mean()
median=salaries.median()
min=salaries.min()
max=salaries.max()
range = max - min
std_dev_salary =salaries.std()

print(f"Mean Salary: {mean}")
print(f"Median Salary: {median}")
print(f"Minimum Salary: {min}")
print(f"Maximum Salary: {max}")
print(f"Salary Range: {range}")
print(f"Standard Deviation of Salary: {std_dev_salary}")
```

```python
df['JobTitle'].interpolate(inplace=True)
df['BasePay'].interpolate(inplace=True)
df['OvertimePay'].interpolate(inplace=True)
df['OtherPay'].interpolate(inplace=True)
df['TotalPay'].interpolate(inplace=True)
df['Agency'].interpolate(inplace=True)
df['Year'].interpolate(inplace=True)
df.fillna({'Notes':0,'Status':0,'Benefits':0})

missing_values = df.isnull().sum()
print("\nMissing values in each column:")
print(missing_values)
```

```python
import matplotlib.pyplot as plt
plt.hist(df['TotalPay'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()


department_counts = df['JobTitle'].value_counts()
plt.bar(department_counts.index, department_counts.values, color='skyblue')
plt.title('Proportion of Employees in Different Departments')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
Grouped_Analysis = df.groupby('JobTitle')['TotalPay'].describe()
print(Grouped_Analysis)
```

```python
salary = 'TotalPay'
numerical_column = 'Year'
correlation_coefficient = df[salary].corr(df[numerical_column])

print(f"Correlation between {salary} and {numerical_column}: {correlation_coefficient}")

plt.scatter(df[numerical_column], df[salary], alpha=0.5)
plt.xlabel(numerical_column)
plt.ylabel(salary)
plt.title(f'Scatter Plot: {salary} vs. {numerical_column}')
plt.show()
```

In the first part, I used df.shape to return the number of rows and columns and df.dtypes to return the data type for each column and df.isnull to checked for missing values in each column .
In the second part ,I calculate the basic statistics .
In third part , the choice of data cleaning method depends on the nature of the data and the analysis goals , so I use interpolation where missing values can be estimated based on the values before and after the missing point and I used fillna methode for some column to ensure the dataset is suitable for further analysis.
In the fourth part , create two figure one histogram for the distribution of salaries using plt.hist , in seconde figure I used plt.bar to create a bar chart for the proportion of employees in different departments(JobTitle) and I put a name for each axis in two figures .
In the fifth part , I used groupby function in pandas to group the data by the 'JobTitle' column and calculate the basic statistics .
In last part , investigated the correlation between salary'TotalPay' and 'year ', finding correlation coefficient indicating nature of correlation and used plt.scatter to create a scatter plot, where the x-axis represents 'Year', and the y-axis represents salary.
